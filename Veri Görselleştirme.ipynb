# Veri Görselleştirme

```python
import seaborn as sns
import matplotlib.pyplot as plt

# İris veri setini yükleme
iris = sns.load_dataset("iris")

# Yoğunluk grafiği çizimi
sns.kdeplot(iris['sepal_length'], shade=True)

# Grafiği gösterme
plt.show()
```
